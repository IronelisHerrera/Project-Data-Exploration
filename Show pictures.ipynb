# Show pictures

```python
from tensorflow.keras.applications import imagenet_utils
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import os
```

# Create array with our pictures data

```python
rootFolder = 'dataset-images/'
folders = os.listdir(rootFolder)
pictures = []

print('All our folders')
print(folders)

for folder in folders:
    root = rootFolder+ folder+'/'
    currentFolder = os.listdir(root)
    for file in currentFolder:
        rootPicture = root+file
        pictures.append(rootPicture)    
```

# Show pictures from array list

```python
count = 0
print("Total pictures: ", len(pictures))
for pictureUrl in pictures:
    img = image.load_img(pictureUrl, target_size=(224,224))
    plt.figure()
    plt.suptitle(pictureUrl)
    plt.imshow(img)
    count = count + 1
    if(count >= 10):
        break
```
